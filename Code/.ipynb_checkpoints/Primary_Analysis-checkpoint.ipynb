Open the data files and load the data

Correlation analysis of the input parameters with the output parameters
=======================================================================

1. Chemistry data
-----------------

```python
chemistry_inputs = chemistry_data['input_data_mean_normalized']
chemistry_targets = chemistry_data['target_data']

num_input_params = chemistry_inputs.shape[1]
num_target_fields = chemistry_targets.shape[1]

for field in range(num_target_fields):
    target_seq = chemistry_targets[:,field]
    correlation_coefficients = []
    for input_param in range(num_input_params):
        input_seq = chemistry_inputs[:,input_param]
        corr_coeff = np.corrcoef(input_seq, target_seq)[0,1]
        correlation_coefficients.append(corr_coeff)
    plt.figure()
    plt.bar(correlation_coefficients)
    plt.show()
```
